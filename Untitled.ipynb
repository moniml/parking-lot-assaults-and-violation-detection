```python
import yaml
from ultralytics import YOLO

# The 'data' dictionary is already defined in a previous cell with the correct paths.
# We need to write this dictionary to the data.yaml file to ensure correct paths are used.
with open('/content/dataset/data.yaml', 'w') as f:
    yaml.dump(data, f, default_flow_style=False)

model = YOLO("yolov8n.pt")

model.train(
    data="/content/dataset/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16
)
```
